```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

options = Options()
options.headless = True
options.add_argument("--window-size=1920,1200")

driver = webdriver.Chrome(options=options, executable_path=DRIVER_PATH)
driver.get("https://www.tradingview.com/markets/stocks-usa/sectorandindustry-industry/biotechnology/")
elem = driver.find_element_by_xpath("//span[@class='tv-load-more__btn']")
while(elem.is_displayed()):
    driver.execute_script("arguments[0].click();", elem)
    time.sleep(5)
#driver.execute_script("arguments[0].click();", driver.find_element_by_xpath("//span[@class='tv-load-more__btn']"))
#time.sleep(5)

soup = BeautifulSoup(driver.page_source, 'html.parser')
ticker_data = []
stockname_data = []

for tr in soup.find_all('tr'):
    #tds = tr.find_all('td')
    ticker = tr.find('a', class_='tv-screener__symbol')
    stock_name = tr.find('span', class_='tv-screener__description')
    if None in (ticker, stock_name):
        continue
    ticker_data.append(ticker.text.strip())
    stockname_data.append(stock_name.text.strip())
    #print(ticker.text.strip())
    #print(stock_name.text.strip())
    #print("\n\n")
driver.quit()
df = pd.DataFrame(ticker_data, stockname_data)
df.to_csv('C:/Users/Shreyas Kulkarni/Desktop/Varshini/Tradingview_advertisement.csv')
```
